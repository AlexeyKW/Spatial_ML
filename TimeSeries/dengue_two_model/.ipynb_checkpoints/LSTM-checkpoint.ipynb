```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from scipy import stats
```

```python
X_train = pd.read_csv('data_dengue/dengue_features_train.csv', index_col=None)
df_y_train = pd.read_csv('data_dengue/dengue_labels_train.csv', index_col=None)
X_test = pd.read_csv('data_dengue/dengue_features_test.csv', index_col=None)
```

```python
X_train['week_start_date'] = pd.to_datetime(X_train['week_start_date'])
X_train['week_start_date'] = X_train['week_start_date'] - pd.to_datetime('1980-01-01')
X_train['week_start_date'] = pd.to_numeric(X_train['week_start_date'])/100000000000
X_train_new = X_train.fillna(X_train.mean())
X_train_sj=X_train_new[X_train_new.city=='sj']
X_train_iq=X_train_new[X_train_new.city=='iq']
```

```python
X_test['week_start_date'] = pd.to_datetime(X_test['week_start_date'])
X_test['week_start_date'] = X_test['week_start_date'] - pd.to_datetime('1980-01-01')
X_test['week_start_date'] = pd.to_numeric(X_test['week_start_date'])/100000000000
X_test_new = X_test.fillna(X_test.mean())
X_test_sj=X_test_new[X_test_new.city=='sj']
X_test_iq=X_test_new[X_test_new.city=='iq']
```

```python
stat, p = stats.normaltest(df_y_train.total_cases)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Data looks Gaussian (fail to reject H0)')
else:
    print('Data does not look Gaussian (reject H0)')
```

```python
sns.distplot(df_y_train.total_cases);
print( 'Kurtosis of normal distribution: {}'.format(stats.kurtosis(df_y_train.total_cases)))
print( 'Skewness of normal distribution: {}'.format(stats.skew(df_y_train.total_cases)))
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
df_y_train['total_cases'].hist(bins=50)
plt.title('Dengue Distribution')
plt.subplot(1,2,2)
stats.probplot(df_y_train['total_cases'], plot=plt);
df1=df_y_train.loc[:,['week_start_date','total_cases']]
df1.describe().T
```

```python
X_train_float = X_train_new.astype('float64')
X_test_float = X_test_new.astype('float64')
```

```python
X_train_float.head()
```

```python
min_max_scaler = MinMaxScaler(feature_range=(0, 1))
X_train_float_scaled = min_max_scaler.fit_transform(X_train_float)
X_test_float_scaled = min_max_scaler.transform(X_test_float)
```

```python
print(X_train_float_scaled[:5, :])
print(X_train_float_scaled.shape)
```

```python
# split into samples (e.g. 1456/182 = 8)
samples = list()
length = 182
# step over the 5,000 in jumps of 200
for i in range(0,1456,length):
    sample = X_train_float_scaled[i:i+length]
    samples.append(sample)
print(len(samples))
```

```python
data = np.array(samples)
print(data.shape)
```

```python
y_train_array = np.array(df_y_train["total_cases"])
samples_y = list()
for i in range(0,1456,length):
    sample = y_train_array[i:i+length]
    samples_y.append(sample)
print(len(samples_y))
```

```python
data_y = np.array(samples_y)
print(data_y.shape)
```

```python
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(182, 25)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(data, data_y, epochs=200, verbose=0)
```

```python
data_1s_test = X_test_float_scaled.reshape(X_test_float_scaled.shape[0],1,X_test_float_scaled.shape[1])
data_1s_test.shape
```

```python
predict = model_1s.predict(data_1s_test)
```

```python
result_columns = X_test[['city','year','weekofyear']]
result_cases = pd.DataFrame(predict)
result_cases.columns = ['total_cases']
result_df = pd.concat([result_columns, result_cases], axis=1)
result_df['total_cases'] = result_df['total_cases'].round().astype(int)
result_df.head()
```
